```python
import pandas as pd
df = pd.read_csv("./data/robots2.csv")
```

```python
start_date = "2017-01-01"
df = df[df['Date']>=start_date]
```

```python
df
```

```python
import tokenizer, runclustering, runsummarization
```

```python
with open('./stopwords_en.txt') as f:
        stopwords = f.readlines()
        stopwords = [x.strip() for x in stopwords]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

#define vectorizer parameters
tf_lemmatizer = TfidfVectorizer(max_df=0.9, max_features=None,
                                min_df=3, stop_words=stopwords,
                                use_idf=True, tokenizer=tokenizer.tokenize_and_lemmatizer, ngram_range=(1,3))

tf_stemmer = TfidfVectorizer(max_df=0.9, max_features=None,
                                min_df=3, stop_words=stopwords,
                                use_idf=True, tokenizer=tokenizer.tokenize_and_stem, ngram_range=(1,3))

tf_tokenizer = TfidfVectorizer(max_df=0.9, max_features=None,
                                min_df=3, stop_words=stopwords,
                                use_idf=True, tokenizer=tokenizer.tokenize_only, ngram_range=(1,3))
```

```python
lemma_X = tf_lemmatizer.fit_transform(df['Contents'])
stemm_X = tf_stemmer.fit_transform(df['Contents'])
token_X = tf_tokenizer.fit_transform(df['Contents'])
```

```python
print lemma_X.shape
print stemm_X.shape
print token_X.shape
```

```python
from sklearn.cluster import KMeans
```

```python
import matplotlib.pyplot as plt
```

```python
cost = []
for i in range(1,len(df)):
    kmeans = KMeans(n_clusters=i, random_state=0).fit(lemma_X)
    cost.append(kmeans.inertia_)
```

```python
plt.plot(range(1,len(df)), cost)
plt.show()
```

```python
import math
n_clusters = int(math.floor(math.sqrt(len(df))))
print n_clusters
print len(df)
```

```python
kmeans_lemma = KMeans(n_clusters=n_clusters, random_state=0).fit(lemma_X)
df['lemma'] = kmeans_lemma.labels_

kmeans_stemm = KMeans(n_clusters=n_clusters, random_state=0).fit(stemm_X)
df['stemm'] = kmeans_stemm.labels_

kmeans_token = KMeans(n_clusters=n_clusters, random_state=0).fit(token_X)
df['token'] = kmeans_token.labels_
```
