```python
import pandas as pd
```

```python
df = pd.read_csv("speech.csv")
```

```python
df = df.dropna()
df
```

```python
from collections import namedtuple
docs = []
analyzedDocument = namedtuple('AnalyzedDocument', 'words tags')
```

```python
for i, row in df.iterrows():
    words = row['Contents'].lower().split()
    docs.append(analyzedDocument(words, row["Title"]))
```

```python
docs
```

```python
from gensim.corpora.wikicorpus import WikiCorpus
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
from pprint import pprint
import multiprocessing
```

```python
pre = Doc2Vec(min_count=0)
pre.scan_vocab(docs)
```

```python
for num in range(0, 20):
    print('min_count: {}, size of vocab: '.format(num), pre.scale_vocab(min_count=num, dry_run=True)['memory']['vocab']/700)
```

```python
cores = multiprocessing.cpu_count()

models = [
    # PV-DBOW 
    Doc2Vec(dm=0, dbow_words=1, size=200, window=8, min_count=3, iter=10, workers=cores),
    # PV-DM w/average
    Doc2Vec(dm=1, dm_mean=1, size=200, window=8, min_count=3, iter =10, workers=cores),
]
```

```python
models[0].build_vocab(docs)
print(str(models[0]))
models[1].build_vocab(docs)
print(str(models[1]))
```

```python
for model in models:
    %%time model.train(docs)
```
